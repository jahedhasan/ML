```python
#assigning features and label variables
#fist feature
weather = ['Sunny','Sunny','Overcast','Rainy','Rainy','Rainy','Overcast','Sunny','Sunny','Rainy','Sunny','Overcast','Overcast','Rainy']
```

```python
#second feature
temp = ['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild','Cool','Mild','Mild','Mild','Hot','Mild']
```

```python
#label or target variable
play = ['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']
```

```python
#import labelEncoder
from sklearn import preprocessing
```

```python
#creating LabelEncoder
le = preprocessing.LabelEncoder()
#converting string to labels into number
weather_encoded = le.fit_transform(weather) # just for check -
temp_encoded = le.fit_transform(temp) # just for check
label_col = le.fit_transform(play) # just for check
print(weather_encoded)
print(temp)
print(label_col)
```

```python
#combining weather and temp into single listof tuples
features = list(zip(weather_encoded,temp_encoded))
print(features)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3)
```

```python
#train the model using the training sets
model.fit(features,label_col)
```

```python
#predict output
predicted = model.predict([[0,2]]) #0:overcast,2:mild
print(predicted)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))

# Display weather_encoded and temp_encoded data with labels
plt.scatter
plt.scatter(weather_encoded, temp_encoded, c=label_col, cmap='viridis', s=100, label="Data Points")
plt.colorbar(label="Play (1) / Not Play (0)")
plt.xlabel("Weather Encoded (Sunny=2, Overcast=0, Rainy=1)")
plt.ylabel("Temperature Encoded (Hot=1, Mild=2, Cool=0)")
plt.title("Weather & Temperature Encoding for KNN Classification")
plt.legend()
plt.grid(True)
plt.show()
```
