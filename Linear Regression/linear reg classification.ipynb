```python
import numpy as np
import matplotlib.pyplot as plt

class SimpleLinearRegression:
    def __init__(self):
        self.m = None
        self.c = None

    def fit(self, X, y):
        X_mean = np.mean(X)
        y_mean = np.mean(y)
        self.m = np.sum((X - X_mean) * (y - y_mean)) / np.sum((X - X_mean) ** 2)
        self.c = y_mean - self.m * X_mean

    def predict(self, X):
        return self.m * X + self.c

    def classify(self, X, y):
        y_pred = self.predict(X)
        return np.where(y > y_pred, '*', '0')


if __name__ == "__main__":
    np.random.seed(0)
    X = np.random.rand(500) * 10
    y = 2 * X + 1 + np.random.randn(1000)
    model = SimpleLinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)
    classes = model.classify(X, y)
    print("Slope (m):", model.m)
    print("Intercept (c):", model.c)
    print("Classifications:", classes)


    plt.figure(figsize=(6, 5))
    plt.scatter(X[classes == '0'], y[classes == '0'], color='red', label='Zero Class (0)', marker='o')
    plt.scatter(X[classes == '*'], y[classes == '*'], color='blue', label='Star Class (*)', marker='*')
    plt.plot(X, y_pred, color='green', linewidth=2, label='Regression Line')
    plt.axhline(0, color='black', linestyle='--', linewidth=0.7)
    plt.title('Simple Linear Regression with Class Symbols')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.legend()
    plt.grid()
    plt.show()
```
