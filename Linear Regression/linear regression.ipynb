```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_path = '/content/mydrive/MyDrive/ML LAB/Linear Regression/cgpa.csv'
data=pd.read_csv(data_path)
data.head(10)
```

```python
print("Dataset Size")
print("Rows {} Colums {}".format(data.shape[0], data.shape[1]))
```

```python
def linear(X, b0, b1):
    return [ b0 + b1 * x for x in  X]
```

```python
# b1- slop
def slope(X, Y):
    x_ = np.mean(X)
    y_ = np.mean(Y)
    rise = sum([(x - x_) * (y - y_) for x, y in zip(X, Y)])
    run = sum([(x - x_) ** 2 for x,y  in  zip (X, Y)])
    return rise / run
```

```python
# b0 - intercept
def intercept(X, Y, b1):
    x_ = np.mean(X)
    y_ = np.mean(Y)
    return y_ - (b1 * x_)
```

```python
final_data = data[['Hour','CGPA']]
predictor = data['Hour']
target = data['CGPA']
```

```python
plt.figure(figsize=(5,5))
plt.title('H_Study vs CGPA')
plt.xlabel('Study(Hour)')
plt.ylabel('CGPA')
plt.scatter(predictor, target, color='red', marker='*')
plt.show()
```

```python
b1 = slope(predictor, target)
b0 = intercept(predictor, target, b1)
predicted = linear(predictor, b0, b1)
print("Slope: ", b1)
print("Intercept: ", b0)

print("Predicted: ", predicted)
```

```python
plt.figure(figsize=(5,5))
plt.title('Predicted values by Linear Regression')
plt.xlabel('Study(Hour)')
plt.ylabel('CGPA')
plt.scatter(predictor, target, color='red', marker='*')
plt.plot(predictor, predicted, color='blue')
plt.show()
```
